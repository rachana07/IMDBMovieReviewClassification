```python
categories = [1,-1]
train_data = pd.read_csv ('train_file.txt', sep="\t")
test_data = pd.read_csv('test_file.txt', sep="\n\n", engine='python')
test_data.review
```

```python
#Vectorize the test data
test_data1=[];
for j in test_data.review:
    test_data1.append(j)
test_data1

```

```python
vectorizer = CountVectorizer()
test_features = vectorizer.fit_transform(test_data1).todense() 
test_features

test_features1 = [];
for f in test_features:
    test_features1.append(euclidean_distances(test_features[0], f) )
test_features1
```

```python
#Text Preprocessing
pre_train = [];
p = 0;
for i in train_data.review:
    ps = PorterStemmer()
    lemmatizer=WordNetLemmatizer()
    input_str = i.lower()  #To lower case
    input_str = re.sub(r'[^\w\s]','',input_str)
    input_str = re.compile(r'<[^>]+>').sub('', input_str) #Removing html tags
    words = word_tokenize(input_str)
    str1 = ''
    for w in words:
        ps.stem(w) #Stemming
        lemmatizer.lemmatize(w) #Lemmatizing
        str1 = str1 + w
        str1 = str1 + " "
    pre_train.append(str1)

pre_train1 = pre_train[:14999]
```

```python
train_features1 = [];
for f in X_train_counts:
    train_features1.append(cosine_similarity(X_train_counts[0], f) )
train_features1
```

```python
train_sentiments = [];
for i in train_data.sentiment:
    train_sentiments.append(i)
```

```python
# transform a count matrix to a normalized tf-idf representation (tf-idf transformer)
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
```

```python
#KNN Implementation
class KNN:
    def __init__(self, k):
        self.k = k
        
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y
        
    def distance(self, X1, X2):
        distance = cosine_similarity(X1, X2)
    
    def predict(self, X_test):
        final_output = []
        for i in range(len(X_test)):
            d = []
            votes = []
            for j in range(len(X_train)):
                dist = cosine_similarity(X_train[j] , X_test[i])
                d.append([dist, j])
            d.sort()
            d = d[0:self.k]
            for d, j in d:
                votes.append(y_train[j])
            ans = Counter(votes).most_common(1)[0][0]
            final_output.append(ans)
            
        return final_output
    
    
    def score(self, X_test, y_test):
        predictions = self.predict(X_test)
        return (predictions == y_test).sum() / len(y_test)
    
    
    
```
